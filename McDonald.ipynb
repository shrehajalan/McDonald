# Preprocessing the dataset

### Conversion from categorical to numerical using LabelEncoder

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
cat = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']
for ele in cat:
    df[ele] = le.fit_transform(df[ele])
df
    
```

```python
scale_mapper = {"I hate it!-5" :1, "-4":2, "-3":3, "-2":4, "-1":5, "0":6, "+1":7, "+2":8, "+3":9, "+4":10, "I love it!+5":11}
df["Like"] = df["Like"].replace(scale_mapper)
```

```python
scale_mapper_2 = {"Once a month":3, "Every three months":2, "Once a year":1, "Once a week": 4, "Never" : 0, "More than once a week":5}
df["VisitFrequency"] = df["VisitFrequency"].replace(scale_mapper_2)
```

```python
scale_mapper_3 = {"Male":1, "Female":0}
df["Gender"] = df["Gender"].replace(scale_mapper_3)
        
```

```python
df.head()
```

### Dropping of similar column

```python
df.drop(columns=['yummy'],inplace=True)
```

### Check Clustering Tendency

```python

hop=hopkins(df_std,150)
print(hop)
###As the value lies between 0- 0.5 the given dataset can be well clustered
```

### Finding best k
### Elbow method to get the best k

```python

distance= []
K = range(1,10)
for num_clusters in K :
 kmeans = KMeans(n_clusters=num_clusters)
 kmeans.fit(df_std)
 distance.append(kmeans.inertia_)
plt.plot(K,distance,'bx-')
plt.xlabel('Values of K')
plt.ylabel('Sum of squared distances/Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

### Applying k-means

```python

kmeans = KMeans(n_clusters=5)
kmeans.fit(df_std)

clusters=kmeans.cluster_centers_
print(clusters)
```

```python
y_pred = kmeans.fit_predict(df_std)
print(y_pred)
```

### Hierarchial clustering 

```python
df_sample=df.sample(frac=0.025)
```

```python
import sys
sys.setrecursionlimit(100000)
import scipy.cluster.hierarchy as sch
dendrogrm = sch.dendrogram(sch.linkage(df_sample, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('placements')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
model.fit(df_sample)
```

```python
y_pred_hier = model.fit_predict(df_sample)
print(y_pred_hier)
```
